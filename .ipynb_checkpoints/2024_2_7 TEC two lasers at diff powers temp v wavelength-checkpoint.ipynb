```python
# list of resistances for each laser and power
res1_15 = [12.001, 14.000, 16.000, 18.002]
res1_20 = [12.001, 14.000, 16.000, 18.001]
res1_25 = [12.000, 14.000, 16.003, 18.001]

res2_15 = [12.001, 14.001, 16.002, 18.002]
res2_20 = [12.000, 14.003, 16.003, 18.002]
res2_25 = [12.000, 14.000, 16.001, 18.001]
```

```python
# Using the thermistor resistance to temperature forumula from Thorlabs TEC manual and specs of HL6756MG laser
```

```python
# nominal temperature in Kelvin
T_o = 298.15

# thermistor nominal resistance at nominal temperature in kOhm
R_o = 10

# energy constant/B value in Kelvin
B = 3977

# append temperatures to a list for each run
temps1_15 = []
temps1_20 = []
temps1_25 = []

temps2_15 = []
temps2_20 = []
temps2_25 = []
```

```python
# Converting resistance to temperature and changing temp in Kelvin to Celcius for each run

for i in range(0, len(wavelengths1_15)):
    T_K = (B * T_o)/((T_o * np.log(res1_15[i]/R_o)) + B)
    T_C = T_K - 273.15
    temps1_15.append(T_C)
    
for i in range(0, len(wavelengths1_20)):
    T_K = (B * T_o)/((T_o * np.log(res1_20[i]/R_o)) + B)
    T_C = T_K - 273.15
    temps1_20.append(T_C)
    
for i in range(0, len(wavelengths1_25)):
    T_K = (B * T_o)/((T_o * np.log(res1_25[i]/R_o)) + B)
    T_C = T_K - 273.15
    temps1_25.append(T_C)
    
for i in range(0, len(wavelengths2_15)):
    T_K = (B * T_o)/((T_o * np.log(res2_15[i]/R_o)) + B)
    T_C = T_K - 273.15
    temps2_15.append(T_C)
    
for i in range(0, len(wavelengths2_20)):
    T_K = (B * T_o)/((T_o * np.log(res2_20[i]/R_o)) + B)
    T_C = T_K - 273.15
    temps2_20.append(T_C)
    
for i in range(0, len(wavelengths2_25)):
    T_K = (B * T_o)/((T_o * np.log(res2_25[i]/R_o)) + B)
    T_C = T_K - 273.15
    temps2_25.append(T_C)
```

```python
# Plotting temperature vs wavelength graph
```

```python
import matplotlib.pyplot as plt
from scipy import stats 
```

```python
# line fit of data
fit1 = stats.linregress(wavelengths1_15, temps1_15)
fit2 = stats.linregress(wavelengths1_20, temps1_20)
fit3 = stats.linregress(wavelengths1_25, temps1_25)

fit4 = stats.linregress(wavelengths2_15, temps2_15)
fit5 = stats.linregress(wavelengths2_20, temps2_20)
fit6 = stats.linregress(wavelengths2_25, temps2_25)
```

```python
plt.plot(wavelengths1_15, temps1_15, 'o', color='blue', label='original data (laser1)')
plt.plot(wavelengths1_15, fit1.intercept + fit1.slope*np.array(wavelengths1_15), 'b', label='fitted line (laser1)')

plt.plot(wavelengths1_20, temps1_20, 'o', color='blue')
plt.plot(wavelengths1_20, fit2.intercept + fit2.slope*np.array(wavelengths1_20), 'b')

plt.plot(wavelengths1_25, temps1_25, 'o', color='blue')
plt.plot(wavelengths1_25, fit3.intercept + fit3.slope*np.array(wavelengths1_25), 'b')

plt.plot(wavelengths2_15, temps2_15, 'o', color='orange', label='original data (laser2)')
plt.plot(wavelengths2_15, fit4.intercept + fit4.slope*np.array(wavelengths2_15), color='orange', label='fitted line (laser2)')

plt.plot(wavelengths2_20, temps2_20, 'o', color='orange')
plt.plot(wavelengths2_20, fit5.intercept + fit5.slope*np.array(wavelengths2_20), color='orange')

plt.plot(wavelengths2_25, temps2_25, 'o', color='orange')
plt.plot(wavelengths2_25, fit6.intercept + fit6.slope*np.array(wavelengths2_25), color='orange')

plt.legend()
plt.xlabel('Wavelength (nm)')
plt.ylabel('Temperature (C)')
plt.grid(linewidth=0.1)
plt.title('Two 670 nm lasers with temperature controller at different powers (2024-2-7)')
# plt.txt? couldn't get it to work
plt.show()
```
